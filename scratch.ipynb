```python
import json
import pandas as pd
```

```python
all_data = json.load(open("public/data.json"))
stats = json.load(open("public/stats.json"))
```

```python
all_data[-1]
```

```python
def fetch_salary(data):
    sal = data.get('Gross (per annum)', 'NA')
    name = data['Company Name'].strip()
    if sal == 'NA':
        print("Fetching CTC (INR) instead of Gross (per annum) for ", name)    
        sal = data.get('CTC (in INR)', -1)
        print(sal)
    
    return sal
```

```python
all_data
```

```python
augmented_data = {}
for idx, data in enumerate(all_data):
    key = f'{data['Company Name'].strip()}_{data['Profile'].strip()}'
    augmented_data[key] = {
        'idx' :  idx,
        'CTC' : fetch_salary(data)
    }
```

```python
augmented_data['ZS Associates Pvt. Ltd.']
```

```python
for i in stats:
    key = f'{i['company_name'].strip()}_{i['profile'].strip()}'
    try:
        
        i['proforma_idx'] = augmented_data[key]['idx']
        i['gross'] = augmented_data[key]['CTC']
    except KeyError:
        i['proforma_idx'] = -1
        i['gross'] = -1

```

```python
json.dump(stats, open("public/stats_augmented.json", "w"), indent=4)
```
